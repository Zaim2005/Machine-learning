```python
#a = pd.get_dummies(d.v1,dtype = float)
a = d['v1'].map({'ham' : 0 , 'spam' : 1})
a
df = pd.concat([d,a],axis = 'columns')
df.columns = ('data','v2','3','4','5','spam')
final = df[['v2','spam']]
```

```python
final
```

```python
X = final['v2']
Y = final['spam']
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.33, random_state=42)
```

```python
tfidf = TfidfVectorizer(min_df = 1 , stop_words='english',lowercase = True)
X_train_features = tfidf.fit_transform(X_train)
X_test_features = tfidf.fit_transform(X_test)
```

```python
print(X_train_features)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train_features,y_train)
```

```python
print(X_test_features)
```

```python
predict = clf.predict(X_train_features)
accuracy = accuracy_score(y_train,predict)
```
